```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
url = 'https://raw.githubusercontent.com/anirudhpnbb/datasets/refs/heads/master/2008_swing_states.csv'
df = pd.read_csv(url)

# Step 3: Create the 'dem_share' column
df['dem_share'] = df['dem_votes'] / df['total_votes']

# Step 4: Create the boxplot
sns.boxplot(data=df, x="state", y="dem_share")
plt.xticks(rotation=90)
plt.title("Democratic Vote Share by State")
plt.ylabel("Democratic Share (0 to 1)")
plt.xlabel("State")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'dem_share' column exists
df['dem_share'] = df['dem_votes'] / df['total_votes']

# Use catplot to group similar dem_share values into bins
sns.catplot(
    data=df,
    x="state",
    y="dem_votes",
    kind="strip",
    height=6,
    aspect=2
)

plt.xticks(rotation=90)
plt.title("Democratic Votes by State (2008)")
plt.xlabel("State")
plt.ylabel("Democratic Votes")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x="dem_share", y="dem_votes")
plt.title("Democratic Votes vs. Democratic Share")
plt.xlabel("Democratic Vote Share")
plt.ylabel("Democratic Votes")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='state', y='dem_votes')
plt.title("Total Democratic Votes by State (2008)")
plt.xlabel("State")
plt.ylabel("Democratic Votes")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=df, x="dem_share", kde=True, bins=30)
plt.title("Distribution of Democratic Vote Share (2008)")
plt.xlabel("Democratic Share")
plt.ylabel("Number of Counties")
plt.show()
```
